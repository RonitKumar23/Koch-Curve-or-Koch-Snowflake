```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
a=[0,0]
b=[30,0]
```

```python
def midpoint(a,b):
    mid1=(((2*a[0]+b[0])/3),((2*a[1]+b[1])/3))
    mid2=(((a[0]+2*b[0])/3),((a[1]+2*b[1])/3))
    return mid1,mid2
```

```python
def rotate(a,b,n,r):
    mid1,mid2=midpoint(a,b)
    x,y=mid1
    ox,oy=mid2
    
    qx = ox + math.cos(r) * (x-ox) - math.sin(r)*(y-oy)
    qy = oy + math.sin(r) * (x-ox) + math.cos(r)*(y-oy)
    
    point=qx,qy
    
    #print(qx,qy)
    if n>1:
        plt.plot([mid1[0],qx],[mid1[1],qy])
        plt.plot([mid2[0],qx],[mid2[1],qy])
        plt.plot([a[0],mid1[0]],[a[1],mid1[1]])
        plt.plot([mid2[0],b[0]],[mid2[1],b[1]])
        plt.plot([mid1[0],mid2[0]],[mid1[1],mid2[1]],color='white')
    
    n=n+1
    if n<5:
        rotate(a,mid1,n,r)
        rotate(mid1,point,n,r)
        rotate(point,mid2,n,r)
        rotate(mid2,b,n,r)
```

```python
rotate(a,b,0,math.radians(60))
rotate(b,[15,26],0,math.radians(60))
rotate([15,26],a,0,math.radians(60))
```
